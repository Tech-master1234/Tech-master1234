```python
%%sql
DELIMITER //

CREATE PROCEDURE if not exists GenerateCustomers()
BEGIN
    DECLARE i INT DEFAULT 1;
    WHILE i <= 1000 DO
        INSERT INTO Customers (first_name, last_name, email, city) VALUES 
        (CONCAT('FirstName', i), CONCAT('LastName', i), CONCAT('customer', i, '@example.com'), CONCAT('City', FLOOR(1 + (RAND() * 100))));
        SET i = i + 1;
    END WHILE;
END //

DELIMITER ;

CALL GenerateCustomers();

DELIMITER //

CREATE PROCEDURE if not exists GenerateLargeCustomerData()
BEGIN
    DECLARE i INT DEFAULT 1;
    WHILE i <= 10000 DO
        INSERT INTO Customers (first_name, last_name, email, city) VALUES
        (CONCAT('FirstName', i), CONCAT('LastName', i), CONCAT('customer', i, '@example.com'), CONCAT('City', FLOOR(1 + (RAND() * 100))));
        SET i = i + 1;
    END WHILE;
END //

DELIMITER ;

-- Call the procedure to generate the data
CALL GenerateLargeCustomerData();
```
